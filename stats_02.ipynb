```python
# Statistics Advance-2 Assignment

# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

"""
PMF (Probability Mass Function):
- Used for discrete random variables
- Gives the probability that a discrete random variable is exactly equal to some value

Example of PMF: Rolling a fair six-sided die
P(X = 1) = P(X = 2) = P(X = 3) = P(X = 4) = P(X = 5) = P(X = 6) = 1/6

PDF (Probability Density Function):
- Used for continuous random variables
- Gives the relative likelihood of a continuous random variable falling within a particular range of values

Example of PDF: The normal distribution
f(x) = (1 / (σ * sqrt(2π))) * e^(-(x-μ)^2 / (2σ^2))
where μ is the mean and σ is the standard deviation
"""

# Plotting an example of PMF (discrete uniform distribution)
x = np.arange(1, 7)
y = np.ones(6) / 6
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.bar(x, y)
plt.title('PMF of Rolling a Fair Die')
plt.xlabel('Outcome')
plt.ylabel('Probability')

# Plotting an example of PDF (normal distribution)
x = np.linspace(-5, 5, 100)
y = stats.norm.pdf(x, 0, 1)
plt.subplot(122)
plt.plot(x, y)
plt.title('PDF of Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.tight_layout()
plt.show()

# Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

"""
CDF (Cumulative Distribution Function):
- Gives the probability that a random variable is less than or equal to a certain value
- For a random variable X, CDF is defined as F(x) = P(X ≤ x)

Example: For a standard normal distribution, CDF gives the area under the curve to the left of a given point.

CDF is used for:
1. Finding probabilities for intervals
2. Generating random numbers from a given distribution
3. Comparing distributions
4. Defining quantiles and percentiles
"""

# Plotting CDF of standard normal distribution
x = np.linspace(-5, 5, 100)
y = stats.norm.cdf(x, 0, 1)
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.title('CDF of Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()

# Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

"""
Examples of normal distribution in real-life:
1. Heights of people in a population
2. IQ scores
3. Measurement errors in scientific experiments
4. Blood pressure readings

Parameters of normal distribution:
1. Mean (μ): Determines the center of the distribution
2. Standard deviation (σ): Determines the spread of the distribution

- Changing μ shifts the distribution left or right
- Increasing σ makes the distribution wider and flatter
- Decreasing σ makes the distribution narrower and taller
"""

# Plotting normal distributions with different parameters
x = np.linspace(-10, 10, 100)
y1 = stats.norm.pdf(x, 0, 1)
y2 = stats.norm.pdf(x, 0, 2)
y3 = stats.norm.pdf(x, -2, 1)

plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='μ=0, σ=1')
plt.plot(x, y2, label='μ=0, σ=2')
plt.plot(x, y3, label='μ=-2, σ=1')
plt.title('Normal Distributions with Different Parameters')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.

"""
Importance of Normal Distribution:
1. Central Limit Theorem: Means of samples from any distribution approximate a normal distribution
2. Many natural phenomena follow normal distribution
3. Basis for many statistical methods and hypothesis tests
4. Useful in modeling and predicting real-world scenarios

Real-life examples:
1. Height and weight distributions in a population
2. Standardized test scores (e.g., SAT, GRE)
3. Measurement errors in manufacturing processes
4. Financial returns in stock markets (approximate)
5. Blood pressure readings in a large population
"""

# Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

"""
Bernoulli Distribution:
- Models a single trial with two possible outcomes (success or failure)
- Has a single parameter p (probability of success)
- PMF: P(X = k) = p^k * (1-p)^(1-k), where k is 0 or 1

Example: Flipping a coin once (Head: success, Tail: failure)

Difference between Bernoulli and Binomial Distribution:
- Bernoulli: Single trial
- Binomial: Multiple independent Bernoulli trials
- Binomial has two parameters: n (number of trials) and p (probability of success)
"""

# Plotting Bernoulli distribution
p = 0.6
x = [0, 1]
y = [1-p, p]
plt.figure(figsize=(8, 5))
plt.bar(x, y)
plt.title(f'Bernoulli Distribution (p={p})')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.show()

# Q6: Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

"""
Given:
μ = 50 (mean)
σ = 10 (standard deviation)
x = 60 (value we're comparing against)

Step 1: Calculate the z-score
z = (x - μ) / σ = (60 - 50) / 10 = 1

Step 2: Find the probability using the standard normal distribution
P(X > 60) = 1 - P(X ≤ 60) = 1 - Φ(z)
where Φ is the CDF of the standard normal distribution
"""

z = (60 - 50) / 10
probability = 1 - stats.norm.cdf(z)

print(f"The probability that a randomly selected observation will be greater than 60 is approximately {probability:.4f}")

# Q7: Explain uniform Distribution with an example.

"""
Uniform Distribution:
- All outcomes are equally likely
- Can be discrete or continuous

Continuous Uniform Distribution:
- PDF is constant over an interval [a, b]
- PDF: f(x) = 1 / (b - a) for a ≤ x ≤ b

Example: Randomly selecting a number between 0 and 1
"""

# Plotting Uniform Distribution
a, b = 0, 1
x = np.linspace(-0.5, 1.5, 100)
y = stats.uniform.pdf(x, a, b-a)
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.title('Uniform Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Density')
plt.ylim(0, 1.5)
plt.show()

# Q8: What is the z score? State the importance of the z score.

"""
Z-score:
- Measures how many standard deviations an observation is from the mean
- Formula: z = (x - μ) / σ
  where x is the observation, μ is the mean, and σ is the standard deviation

Importance of z-score:
1. Standardization: Converts values from different normal distributions to a standard scale
2. Comparison: Allows comparison of values from different normal distributions
3. Probability calculation: Used to find probabilities using standard normal distribution tables
4. Outlier detection: Helps identify unusual values in a dataset
"""

# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

"""
Central Limit Theorem (CLT):
States that the distribution of sample means approximates a normal distribution as the sample size becomes larger, regardless of the population's distribution.

Significance of CLT:
1. Enables statistical inference about population parameters
2. Justifies the use of normal distribution in many statistical methods
3. Allows for the construction of confidence intervals
4. Supports the validity of many hypothesis tests
5. Fundamental to understanding sampling distributions
"""

# Q10: State the assumptions of the Central Limit Theorem.

"""
Assumptions of the Central Limit Theorem:

1. Random Sampling: Samples must be randomly selected from the population
2. Independence: Observations within and between samples must be independent
3. Sample Size: The sample size should be sufficiently large (usually n ≥ 30)
4. Finite Variance: The population should have a finite variance
5. Similar Distribution: For smaller sample sizes, the population should not be extremely skewed
"""
```
